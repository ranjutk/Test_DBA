### Data-2 The goal is to see the home and away game advantage in the 2012-13 English Premier League Season

Let's import the library and dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/footballcsv/eng-england/master/2010s/2012-13/1-premierleague.csv')  #raw file
df.head()
```

We are removing irreleavant columns and changing the column names. We dont need Half Time score and let's rename Team 1 as Home Team and Team 2 as Away Team. 

```python
df=df.drop(['Round','HT'], 1) #deleting Columns
df.columns=['Date','Home Team','FT','Away Team'] #Renaming columns
df.head()
```

**Now we are going to extract the Full time score (FT) and assign it to home team and away team. We will create another column near the home team and away team with individual team scores. Then remove the call FT as it will be redundant**

```python
ft_list=df['FT'].values
leng=len(ft_list) #finding total number of enteries in the column
count =0
home_goal=[]
away_goal=[]
while count<leng:
    home_goal.append(int(ft_list[count][0]))    #adding the first value i.e. the home goal scored into new list
    away_goal.append(int(ft_list[count][2]))    #adding the last value i.e. the away goal scored into new list
    count=count+1
df.insert(loc=2, column='Home Team Score', value=home_goal) #adding the home goal to a new column
df['Away Team Score']= away_goal #last index anyway  #adding the away gaol to a new column
df=df.drop(['FT'],1)  # deleting full time score as it's redundant
df.head()
```

**We have a bit of a problem here as the raw data has the game round number associated with each team. We have to clean it up
to give the corrct team name. Lets remove that so it will be easier to consilidate data**

```python
ht_list=df['Home Team'].values #extracting home team names into a list
at_list=df['Away Team'].values #extracting away team 
home_team=[]
away_team=[]
i=0
j=0
ht_name=''
at_name=''
while i<leng:  #this entire loop is to extract the values until the character '(' appears. This will make the team names uniform
    while j>=0:
            ht_name=ht_name+ht_list[i][j] #adding each extracted characters into an empty string variable
            j=j+1
            if ht_list[i][j]=='(':
                home_team.append(ht_name)
                ht_name=''
                j=0
                break
    while j>=0:
        at_name=at_name+at_list[i][j]
        j=j+1
        if at_list[i][j] =='(':
            away_team.append(at_name)
            at_name=''
            j=0
            break
    i=i+1
df['Home Team']=home_team  #replacing the old values with the new corrected values
df['Away Team']=away_team
df.head()
```

**This will be much easier to analyze. Lets look at the goal difference to see who won the match!**

```python
df['Goal Difference']= df['Home Team Score'] - df['Away Team Score']
df.head()
```

```python
df['Home_Win'] = df['Goal Difference']>0 #Checking for game status. Win, lose or darw
df['Away_Win'] = df['Goal Difference']<0
df['Draw'] = df['Goal Difference'] == 0
df.head()
```

**Now lets find the home and away teams**

```python
tidy = pd.melt(df.reset_index(),
               id_vars=['Date'], value_vars=['Away Team', 'Home Team'],
               value_name='Team')
tidy.head()
```

**Now we will see the Home win and away win percentage**

```python
wins = (
    pd.melt(df.reindex(),
            id_vars=['Date', 'Home_Win'],
            value_name='Team', var_name='home_or_away',
            value_vars=['Home Team', 'Away Team'])
   .assign(win=lambda x: x.Home_Win == (x.home_or_away == 'Home Team'))
   .groupby(['Team', 'home_or_away'])
   .win
   .agg(['sum', 'count', 'mean'])
   .rename(columns=dict(sum='n_wins',
                        count='n_games',
                        mean='win_pct'))
)
wins.head(20)
```
